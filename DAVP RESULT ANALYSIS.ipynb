```python
plt.figure(figsize=(12,8))
plt.pie(batch1.RESULT.value_counts(),explode=[0.1,0.1],autopct='%.1f%%',shadow=True,labels=['pass','fail'])
plt.axis('equal')
plt.title('pieplot')
plt.show()   
```

```python
plt.figure(figsize=(12,8))
plt.pie(batch2.RESULT.value_counts(),explode=[0.1,0.1],autopct='%.1f%%',shadow=True,labels=['pass','fail'])
plt.axis('equal')
plt.title('pieplot')
plt.show()  
```
